```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df.drop(columns=['car_ID', 'CarName', 'symboling', 'carheight', 'boreratio', 'stroke', 'compressionratio', 'peakrpm', 'fuelsystem'], inplace=True, errors='ignore')
```

```python
median_price = df['price'].median()
df['price_class'] = (df['price'] > median_price).astype(int)

df.drop('price', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
cat_cols = df.select_dtypes(include=['object']).columns
le = LabelEncoder()
for col in cat_cols:
    df[col] = le.fit_transform(df[col])

print("\nEncoded categorical columns:", list(cat_cols))
```

```python
plt.figure(figsize=(14,10))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
x = df.drop('price_class', axis=1)
y = df['price_class']
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
sc.fit(x)
```

```python
x_scaled = sc.transform(x)
```

```python
x_scaled
```

```python
df['price_class'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled,y,
                                                   train_size = 0.8,
                                                   random_state = 42,
                                                   stratify = y)
```

```python
print(x_train.shape, x_test.shape)
print(y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression()
```

```python
log_reg.fit(x_train, y_train)
```

```python
log_reg.score(x_train, y_train)
```

```python
y_pred = log_reg.predict(x_test) 
print(len(y_pred))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
accuracy_score(y_pred, y_test) 
```

```python
confusion_matrix(y_pred, y_test)
```

```python
sns.heatmap(confusion_matrix(y_pred, y_test), annot = True,cmap = 'Reds', fmt = '.2f')
plt.show()
```

```python
print(classification_report(y_pred, y_test))
```

```python
import joblib
```

```python
joblib.dump(log_reg, "car_price_classifier.pkl")
joblib.dump(sc, "scaler.pkl")

print("✅ Model and Scaler saved successfully!")
```
